# tushare

Tushare是免费的数据平台。有两个版本：[tushare.org](http://tushare.org/)(旧版)和[tushare.pro](https://tushare.pro/)(新版)。旧版是开源的数据收集软件，新版数据存储在tushare自己的服务器上。

由于后期tushare以新版为主，所以dfdata中的数据源tushare指tushare.pro(新版)。
